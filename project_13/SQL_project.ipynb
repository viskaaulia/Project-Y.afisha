###  Jumlah ulasan pengguna dan rating rata-rata untuk setiap buku

- **Jumlah ulasan pengguna untuk setiap buku**

```python
query = '''SElECT 
                books.title,
                books.book_id,
                COUNT(DISTINCT review_id) as total_review
                
            FROM
                books
                LEFT JOIN reviews on books.book_id = reviews.book_id
                GROUP BY
                    books.book_id
                ORDER BY
                    total_review DESC
                LIMIT 10
                '''
total_review = pd.io.sql.read_sql(query, con = engine)
total_review
```

Buku yang memiliki ulasan pengguna paling tinggi yaitu buku dengan judul Twilight (Twilight #1) yaitu sebanyak 7 ulasan

- **Rating rata-rata untuk setiap buku**

```python
query = '''SElECT 
                books.title,
                books.book_id,
                COUNT (DISTINCT review_id) as total_review,
                AVG (rating) as avg_rating
                
            FROM
                books
                LEFT JOIN reviews on books.book_id = reviews.book_id
                LEFT JOIN ratings on  books.book_id = ratings.book_id
                GROUP BY
                    books.book_id
                ORDER BY
                    avg_rating DESC,
                    total_review DESC
                LIMIT 10
                '''
avg_rating_and_review = pd.io.sql.read_sql(query, con = engine)
avg_rating_and_review
```

Buku yang memiliki rating tertinggi dengan jumlah review tertinggi yaitu buku dengan judul A Dirty Job (Grim Reaper #1), yaitu rating dengan niai 5 dan jumlah ulasan sebannyak 4 ulasan.

###  Penerbit yang telah menerbitkan jumlah buku terbanyak

```python
query = '''SELECT
                publishers.publisher_id,
                publishers.publisher,
                COUNT(DISTINCT books.title) as total_book
            FROM
                books
                LEFT JOIN publishers on books.publisher_id = publishers.publisher_id
            WHERE
                books.num_pages > 50
            GROUP BY
                publishers.publisher, publishers.publisher_id
            ORDER BY
                total_book DESC
            LIMIT 10
                '''
publisers_ = pd.io.sql.read_sql(query, con = engine)
publisers_
```

Penerbit yang telah menerbitkan jumlah buku paling banyak dengan jumlah halaman lebih dari 50 halaman adalah penerbit Penguin Books

###  Penulis dengan rating rata-rata buku tertinggi

```python
query = '''SELECT
                authors.author,
                AVG (subquery_2.avg_rating) as final_rating
            FROM (
                SELECT
                    books.title,
                    books.author_id,
                    subquery_1.avg_rating
                FROM
                    (SELECT
                        book_id,
                        COUNT (rating_id) AS rating_cnt,
                        AVG (rating) AS avg_rating
                    FROM
                        ratings
                    GROUP BY
                        book_id
                    HAVING
                        COUNT(rating_id) > 50) AS subquery_1
                    INNER JOIN books ON books.book_id = subquery_1.book_id) AS subquery_2
                INNER JOIN authors ON authors.author_id = subquery_2.author_id
            GROUP BY
                author
            ORDER BY
                final_rating DESC
            LIMIT 10
                '''
avg_rate = pd.io.sql.read_sql(query, con = engine)
avg_rate
```

Penulis dengan rating rata-rata buku tertinggi yang jumlah ratingnya minimal 50 yaitu penulis J.K. Rowling/Mary GrandPré

### Jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku.

```python
query = '''SELECT
                AVG(Sub.cnt_reviews) AS avg_review
            FROM (
                SELECT
                    ratings.username,
                    COUNT(DISTINCT ratings.book_id) AS cnt_rating,
                    COUNT(DISTINCT reviews.review_id) AS cnt_reviews
                FROM
                    ratings
                LEFT OUTER JOIN reviews ON ratings.username=reviews.username and ratings.book_id=reviews.book_id
            GROUP BY
                ratings.username
            HAVING
                COUNT(DISTINCT ratings.book_id)>50) AS Sub'''

avg_rate = pd.io.sql.read_sql(query, con = engine)
avg_rate
```

Pengguna yang memberi rating terhadap penulis yang menerbitkan buku lebih dari 50 buku, jumlah rata-rata ulasannya yaitu sekitar 25 ulasan.
